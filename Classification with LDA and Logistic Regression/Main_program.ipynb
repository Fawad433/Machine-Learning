## Review: 

## Linear Discriminant Analysis

So far we have only been applying QDA.  Let's write a discLDA function and see if this classifier, which assumes all classes have the same covariance matrix, does better than QDA on our Parkinson's data.

Above we showed that if we assume the same covariance matrix, $\Sigmav$, for each class, where 
$$
\begin{align*}
\Sigmav = \sum_{k=1}^K \frac{N_k}{N} \Sigmav_k,
\end{align*}
$$
our discriminant function becomes
$$
      \begin{align*}
        \delta_k(\xv) = \xv^T \Sigmav^{-1} \muv_k - \frac{1}{2}\muv_k^T \Sigmav^{-1} \muv_k + \log
        P(C=k)
      \end{align*}
$$

## Classification with Nonlinear Logistic Regression Using Neural Networks

For NonLinear Logistic Regression we use this diagram
<img src="http://www.cs.colostate.edu/~anderson/cs480/notebooks/nntwolayerlogisticregression.png">

### Linear Version
General gradient

$$
      \begin{align*}
        \frac{\partial LL(\Wv)}{\partial \Wv_{d,j}} & =  \sum_{n=1}^N \sum_{k=1}^{K} (t_{n,k} - g_{n,k}) \frac{\partial y_{n,k}}{\partial \Wv_{d,j}}
      \end{align*}
$$

For linear logistic regression, $y_{n,j} = \xv_n^T \Wv_{*,j}$, so $\frac{\partial y_{n,k}}{\partial \Wv_{d,j}}$ exists only when $j=k$, so

$$
      \begin{align*}
        \frac{\partial LL(\Wv)}{\partial \Wv_{d,j}} & =  \sum_{n=1}^N (t_{n,j} - g_{n,j}) \frac{\partial y_{n,j}}{\partial \Wv_{d,j}}\\
         & =  \sum_{n=1}^N \left ( t_{n,j} - g_{n,j} \right ) \xv_{d,j}
      \end{align*}
$$

### NonLinear Version

$$
      \begin{align*}
        \Zv &= h(\tilde{\Xv} \Vv)\\
        \Yv &= \tilde{\Zv} \Wv\\
        \Fv &= e^{\Yv}\\
        \Sv &= \Fv \ones{K}\;\;\;\;\;\;\;\;\;\;\;\;\; \text{ sum across columns}\\
        \Gv &= \Fv / \left [ \Sv, \Sv,\ldots,\Sv \right ] \;\;\; \Sv \text{ are column vectors }\\
        LL &=  \sum \Tiv \log \Gv\\
        \grad_\Vv LL &=    \tilde{\Xv}^T \left ( (\Tiv - \Gv) \hat{\Wv}^T \cdot  (1-\Zv^2) \right )\\
        \grad_\Wv LL &=   \tilde{\Zv}^T (\Tiv - \Gv)
      \end{align*}
$$

## Required Code

following are the functions that train and evaluate LDA and neural network logistic regression models.

* `model = trainLDA(X,T,parameters)`
* `percentCorrect = evaluateLDA(model,X,T)`
* `model = trainNN(X,T,parameters)`
* `percentCorrect = evaluateNN(model,X,T)`
The `parameters` argument for `trainNN` is a list of the hidden layers structure and the number of SCG iterations, as in the previous assignment. The value of the `parameters` argument for `trainLDA` is not used.

`trainValidateTestKFoldsClassification` function in `mlutils.py` is used.

The `NeuralNetworkClassifier` class in the above `neuralnetworks.py` file allows to specify 0 hidden layers which is the implementation of Linear Logistic Regression.

```python
import numpy as np
import mlutils as ml
import neuralnetworks as nn
%matplotlib inline
import matplotlib.pyplot as plt
from imp import reload
import qdalda as ql
```

```python
def trainLDA(X,T,parameters = 0):
    lda = ql.LDA()
    lda.train(X,T)
    return lda
```

```python
def evaluateLDA(model,X,T):
    predictedClass,_,_ = model.use(X)   
    return np.sum(predictedClass==T)/X.shape[0] * 100
```

```python
def trainNN(X,T,parameters):
    classes,counts = np.unique(T,return_counts=True)
    nnet = nn.NeuralNetworkClassifier(X.shape[1],parameters[0],len(classes))
    output = nnet.train(X,T,parameters[1], verbose=False)
    return output
```

```python
def evaluateNN(model,X,T):
    c,Pr,_ = model.use(X,allOutputs=True)    
    return np.sum(c==T)/X.shape[0] * 100
```

### Code

First we grab the data and catagorize it into the classes then we seperate it into training, test and validation folds to apply LDA, Linear and NonLinear Logistic Regression. After applying all the methods we compare the results and analyse the accuracy of different methods

```python
def makeMPGData(filename='auto-mpg.data'):
    def missingIsNan(s):
        return np.nan if s == b'?' else float(s)
    data = np.loadtxt(filename, usecols=range(8), converters={3: missingIsNan})
    print("Read",data.shape[0],"rows and",data.shape[1],"columns from",filename)
    goodRowsMask = np.isnan(data).sum(axis=1) == 0
    data = data[goodRowsMask,:]
    print("After removing rows containing question marks, data has",data.shape[0],"rows and",data.shape[1],"columns.")
    X = data[:,1:]
    T = data[:,0:1]
    Xnames =  ['cylinders','displacement','horsepower','weight','acceleration','year','origin']
    Tname = 'mpg'
    return X,T,Xnames,Tname
```

```python
def makeMPGClasses(T):
    bounds = np.arange(5,45,10)
    Tclasses = -np.ones(T.shape).astype(np.int)
    for i,mpg in enumerate(T):
        for k in range(len(bounds)-1):
            if bounds[k] < mpg <= bounds[k+1]:
                Tclasses[i] = bounds[k+1]
        if Tclasses[i] == -1:
            Tclasses[i] = 50  # max mpg is 46.6
    return Tclasses
```

```python
def printResults(label,results):
    print('{:4s} {:>20s}{:>8s}{:>8s}{:>8s}'.format('Algo','Parameters','TrnAcc','ValAcc','TesAcc'))
    print('-------------------------------------------------')
    for row in results:
        # 20 is expected maximum number of characters in printed parameter value list
        print('{:>4s} {:>20s} {:7.2f} {:7.2f} {:7.2f}'.format(label,str(row[0]),*row[1:]))
```

```python
X,T,Xnames,Tname = makeMPGData('auto-mpg.data')
Tclasses = makeMPGClasses(T)
classes,counts = np.unique(Tclasses,return_counts=True)
```

#### Code to Analyse 
This program take random values for test and training folds and apply LDA.

```python
import pdb
def discLDA(X, means,stds, mu, Sigma, prior):
    X = (X-means)/stds
    if Sigma.size == 1:
        Sigma = np.asarray(Sigma).reshape((1,1))
    det = np.linalg.det(Sigma)        
    # if det == 0:
    #    raise np.linalg.LinAlgError('discQDA(): Singular covariance matrix')
    SigmaInv = np.linalg.pinv(Sigma)     # pinv in case Sigma is singular
    mu = mu.reshape((-1,1)) # make mu a column vector
    # pdb.set_trace()
    return np.dot(np.dot(X,SigmaInv), mu) - 0.5 * np.dot(np.dot(mu.T,SigmaInv), mu) + np.log(prior)
def percentCorrect(p, t):
    return np.sum(p.ravel()==t.ravel()) / float(len(t)) * 100

firstI,_ = np.where(Tclasses == 15)
secondI,_ = np.where(Tclasses == 25)
thirdI,_ = np.where(Tclasses == 35)
fourthI,_ = np.where(Tclasses == 50)
firstI = np.random.permutation(firstI)
secondI = np.random.permutation(secondI)
thirdI = np.random.permutation(thirdI)
fourthI = np.random.permutation(fourthI)
trainFraction = 0.8    
nfirst = round(trainFraction*len(firstI))
nsecond = round(trainFraction*len(secondI))
nthird = round(trainFraction*len(thirdI))
nfourth = round(trainFraction*len(fourthI))
    
rowsTrain = np.hstack((firstI[:nfirst], secondI[:nsecond],thirdI[:nthird],fourthI[:nfourth]))
Xtrain = X[rowsTrain, :]
Ttrain = Tclasses[rowsTrain, :]
rowsTest = np.hstack((firstI[nfirst:], secondI[nsecond:], thirdI[nthird:], fourthI[nfourth:]))
Xtest =  X[rowsTest, :]
Ttest =  Tclasses[rowsTest, :]

means,stds = np.mean(Xtrain,0), np.std(Xtrain,0)
Xtrains = (Xtrain-means)/stds

Ttr = (Ttrain==15).reshape((-1))
mu1 = np.mean(Xtrains[Ttr, :],axis=0)
cov1 = np.cov(Xtrains[Ttr, :].T)
Ttr = (Ttrain.ravel()==25).reshape((-1))
mu2 = np.mean(Xtrains[Ttr, :],axis=0)
cov2 = np.cov(Xtrains[Ttr, :].T)
Ttr = (Ttrain.ravel()==35).reshape((-1))
mu3 = np.mean(Xtrains[Ttr, :],axis=0)
cov3 = np.cov(Xtrains[Ttr, :].T)
Ttr = (Ttrain.ravel()==50).reshape((-1))
mu4 = np.mean(Xtrains[Ttr, :],axis=0)
cov4 = np.cov(Xtrains[Ttr, :].T)

covMean = ((cov1 * nfirst) + (cov2 * nsecond) + (cov3 * nthird) +(cov4 * nfourth)) / (nfirst+nsecond+nthird+nfourth)
d1 = discLDA(Xtrain, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
d2 = discLDA(Xtrain, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
d3 = discLDA(Xtrain, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird+nfourth))
d4 = discLDA(Xtrain, means, stds, mu4, covMean, float(nfourth)/(nfirst+nsecond+nthird+nfourth))
predictedTrain = np.argmax(np.hstack((d1, d2, d3, d4)), axis=1)

d1t = discLDA(Xtest, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
d2t = discLDA(Xtest, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
d3t = discLDA(Xtest, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird+nfourth))
d4t = discLDA(Xtest, means, stds, mu4, covMean, float(nfourth)/(nfirst+nsecond+nthird+nfourth))
predictedTest = np.argmax(np.hstack((d1t, d2t, d3t,d4t)), axis=1)

for i in range(len(predictedTrain)):
    if predictedTrain[i] == 0:
        predictedTrain[i] = 15
    if predictedTrain[i] == 1:
        predictedTrain[i] = 25
    if predictedTrain[i] == 2:
        predictedTrain[i] = 35
    if predictedTrain[i] == 3:
        predictedTrain[i] = 50
for i in range(len(predictedTest)):
    if predictedTest[i] == 0:
        predictedTest[i] = 15
    if predictedTest[i] == 1:
        predictedTest[i] = 25
    if predictedTest[i] == 2:
        predictedTest[i] = 35
    if predictedTest[i] == 3:
        predictedTest[i] = 50

plt.plot(Ttrain, 'o-')
plt.plot(predictedTrain, 'o-')
plt.xlabel('samples')
plt.ylabel('class')
print('LDA Percent correct: Train', percentCorrect(predictedTrain, Ttrain), 'Test', percentCorrect(predictedTest,Ttest))



```

In this diagram,one can see the samples which are not properly classified.

1. The accuracy of any new data depends on the versatility of the data used to train data. 
2. If the train and validation data has high accuracy that doesn't mean that the test will give high accuracy
3. Sometimes data with high training accuracy and low validation accuracy can give high Test Accuracy as it all depends on the data used for training

#### Code to analyse 
In this folds for training and test are randomly picked and then we implement the LDA to ree the results.

```python
firstI,_ = np.where(T == 1)
secondI,_ = np.where(T == 2)
thirdI,_ = np.where(T == 3)

firstI = np.random.permutation(firstI)
secondI = np.random.permutation(secondI)
thirdI = np.random.permutation(thirdI)

trainFraction = 0.8    
nfirst = round(trainFraction*len(firstI))
nsecond = round(trainFraction*len(secondI))
nthird = round(trainFraction*len(thirdI))

    
rowsTrain = np.hstack((firstI[:nfirst], secondI[:nsecond],thirdI[:nthird]))
Xtrain = X[rowsTrain, :]
Ttrain = T[rowsTrain, :]
rowsTest = np.hstack((firstI[nfirst:], secondI[nsecond:], thirdI[nthird:]))
Xtest =  X[rowsTest, :]
Ttest =  T[rowsTest, :]

means,stds = np.mean(Xtrain,0), np.std(Xtrain,0)
Xtrains = (Xtrain-means)/stds

Ttr = (Ttrain==1).reshape((-1))
mu1 = np.mean(Xtrains[Ttr, :],axis=0)
cov1 = np.cov(Xtrains[Ttr, :].T)
Ttr = (Ttrain.ravel()==2).reshape((-1))
mu2 = np.mean(Xtrains[Ttr, :],axis=0)
cov2 = np.cov(Xtrains[Ttr, :].T)
Ttr = (Ttrain.ravel()==3).reshape((-1))
mu3 = np.mean(Xtrains[Ttr, :],axis=0)
cov3 = np.cov(Xtrains[Ttr, :].T)


covMean = ((cov1 * nfirst) + (cov2 * nsecond) + (cov3 * nthird)) / (nfirst+nsecond+nthird)
d1 = discLDA(Xtrain, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird))
d2 = discLDA(Xtrain, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird))
d3 = discLDA(Xtrain, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird))

predictedTrain = np.argmax(np.hstack((d1, d2, d3)), axis=1)

d1t = discLDA(Xtest, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird))
d2t = discLDA(Xtest, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird))
d3t = discLDA(Xtest, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird))

predictedTest = np.argmax(np.hstack((d1t, d2t, d3t)), axis=1)

for i in range(len(predictedTrain)):
    if predictedTrain[i] == 2:
        predictedTrain[i] = 3
for i in range(len(predictedTrain)):
    if predictedTrain[i] == 1:
        predictedTrain[i] = 2
for i in range(len(predictedTrain)):
    if predictedTrain[i] == 0:
        predictedTrain[i] = 1
for i in range(len(predictedTest)):
    if predictedTest[i] == 2:
        predictedTest[i] = 3
for i in range(len(predictedTest)):
    if predictedTest[i] == 1:
        predictedTest[i] = 2
for i in range(len(predictedTest)):
    if predictedTest[i] == 0:
        predictedTest[i] = 1
print('LDA Percent correct: Train', percentCorrect(predictedTrain, Ttrain), 'Test', percentCorrect(predictedTest,Ttest))


plt.plot(Ttrain, 'o-')
plt.plot(predictedTrain, 'o-')
plt.xlabel('samples')
plt.ylabel('class')
def percentCorrect(p, t):
    return np.sum(p.ravel()==t.ravel()) / float(len(t)) * 100
```
