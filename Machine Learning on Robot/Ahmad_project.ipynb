## Specifications of DATA

This data is taken from http://mlr.cs.umass.edu/ml/datasets/Wall-Following+Robot+Navigation+Data . I am using these methods to analyze the type of data and to determine which method will best fit the data because In real time when it comes to the readings of the sensors of the robot, every decision matters. So I have applied all the methods on the different set of sensor readings to look for the best-fitted algorithm for this scenario, specially for the task of the wall following robot and for ultrasonic sensors in general.
The data used in this project is being collected from Wall-Following mobile robot SCITOS-G5 and it has following traits to consider before it's being used. 
1. The data was collected as the SCITOS G5 navigates through the room following the wall in a clockwise direction, for 4 rounds. To navigate, the robot uses 24 ultrasound sensors arranged circularly around its "waist". The numbering of the ultrasound sensors starts at the front of the robot and increases in a clockwise direction. 
2. The data file comprises of three different data sets. The $first$ $data$ $file$ contain the values of the measurements of all 24 ultrasound sensors and the corresponding class label. Sensor readings are sampled at a rate of $ 9$ $samples$ $per$ $second$. The $second$ $data$ $file$  contains four sensor readings named 'simplified distances' and the corresponding class label.These simplified distances are referred to as the 'front distance', 'left distance', 'right distance' and 'back distance'.They consist, respectively, of the minimum sensor readings among those within 60-degree arcs located at the front, left, right and back parts of the robot. The $third$ $data$ $file$  contains only the front and left simplified distances and the corresponding class label.
3. One of the most important points is that these 24 ultrasound readings and the simplified distances were collected at the same time step, so each file has the same number of rows (one for each sampling time step). 
4. The wall-following task and data gathering were designed to test the hypothesis that this apparently simple navigation task is indeed a non-linearly separable classification task. So according to the data provider, this data will not give good results with the linear classification methods which I will also analyze in this project.
5. Files with different number of sensor readings were built in order to evaluate the performance of the classifiers with respect to the number of inputs

```python
import numpy as np
import mlutils as ml
import neuralnetworks as nn
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from imp import reload
import qdalda as ql
from scipy.stats import norm
import pdb
```

```python
def discLDA(X, means,stds, mu, Sigma, prior):
    X = (X-means)/stds
    if Sigma.size == 1:
        Sigma = np.asarray(Sigma).reshape((1,1))
    det = np.linalg.det(Sigma)        
    # if det == 0:
    #    raise np.linalg.LinAlgError('discQDA(): Singular covariance matrix')
    SigmaInv = np.linalg.pinv(Sigma)     # pinv in case Sigma is singular
    mu = mu.reshape((-1,1)) # make mu a column vector
    # pdb.set_trace()
    return np.dot(np.dot(X,SigmaInv), mu) - 0.5 * np.dot(np.dot(mu.T,SigmaInv), mu) + np.log(prior)
def percentCorrect(p, t):
    return np.sum(p.ravel()==t.ravel()) / float(len(t)) * 100
```

```python
def discQDA(X, means, stds, mu, Sigma, prior):
    Xc = (X - means) / stds - mu
    if Sigma.size == 1:
        Sigma = np.asarray(Sigma).reshape((1,1))
    det = np.linalg.det(Sigma)        
    # if det == 0:
    #   raise np.linalg.LinAlgError('discQDA(): Singular covariance matrix')
    SigmaInv = np.linalg.pinv(Sigma)     # pinv in case Sigma is singular
    return -0.5 * np.log(det) \
           - 0.5 * np.sum(np.dot(Xc, SigmaInv) * Xc, axis=1).reshape((-1,1)) \
           + np.log(prior)
```

```python
def percentCorrect(p, t):
    return np.sum(p.ravel()==t.ravel()) / float(len(t)) * 100
```

```python
def testcode(filename,Xcolumns,Tcolumn,Type):
    # Type is to identify the method , 1 for LDA else  QDA
    f = open(filename,"r")
    data = np.loadtxt(f ,delimiter=',',usecols=np.arange(Xcolumns))
    f = open(filename,"r")
    T = np.loadtxt(f ,delimiter=',',usecols= Tcolumn, dtype=np.str).reshape((-1, 1))
    P = np.zeros(len(T)).reshape((-1, 1))
    for i in range(len(T)):  
        T[i] = commandCheck(T[i])
        P[i] = int(np.asscalar(T[i]))
    T = P
    X = data
    firstI,_ = np.where(T == 1)
    secondI,_ = np.where(T == 2)
    thirdI,_ = np.where(T == 3)
    fourthI,_ = np.where(T == 4)

    firstI = np.random.permutation(firstI)
    secondI = np.random.permutation(secondI)
    thirdI = np.random.permutation(thirdI)
    fourthI = np.random.permutation(fourthI)

    trainFraction = 0.8    
    nfirst = round(trainFraction*len(firstI))
    nsecond = round(trainFraction*len(secondI))
    nthird = round(trainFraction*len(thirdI))
    nfourth = round(trainFraction*len(fourthI))

    rowsTrain = np.hstack((firstI[:nfirst], secondI[:nsecond],thirdI[:nthird], fourthI[:nfourth]))
    Xtrain = X[rowsTrain, :]
    Ttrain = T[rowsTrain, :]
    rowsTest = np.hstack((firstI[nfirst:], secondI[nsecond:], thirdI[nthird:], fourthI[nfourth:]))
    Xtest =  X[rowsTest, :]
    Ttest =  T[rowsTest, :]        
    
    means,stds = np.mean(Xtrain,0), np.std(Xtrain,0)
    Xtrains = (Xtrain-means)/stds

    Ttr = (Ttrain==1).reshape((-1))
    mu1 = np.mean(Xtrains[Ttr, :],axis=0)
    cov1 = np.cov(Xtrains[Ttr, :].T)
    Ttr = (Ttrain.ravel()==2).reshape((-1))
    mu2 = np.mean(Xtrains[Ttr, :],axis=0)
    cov2 = np.cov(Xtrains[Ttr, :].T)
    Ttr = (Ttrain.ravel()==3).reshape((-1))
    mu3 = np.mean(Xtrains[Ttr, :],axis=0)
    cov3 = np.cov(Xtrains[Ttr, :].T)
    Ttr = (Ttrain.ravel()==4).reshape((-1))
    mu4 = np.mean(Xtrains[Ttr, :],axis=0)
    cov4 = np.cov(Xtrains[Ttr, :].T)
    
    if Type == 1:
        covMean = ((cov1 * nfirst) + (cov2 * nsecond) + (cov3 * nthird)+ (cov4 * nfourth)) / (nfirst+nsecond+nthird+nfourth)
        d1 = discLDA(Xtrain, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
        d2 = discLDA(Xtrain, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
        d3 = discLDA(Xtrain, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird+nfourth))
        d4 = discLDA(Xtrain, means, stds, mu4, covMean, float(nfourth)/(nfirst+nsecond+nthird+nfourth))

        predictedTrain = np.argmax(np.hstack((d1, d2, d3, d4)), axis=1)

        d1t = discLDA(Xtest, means, stds, mu1, covMean, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
        d2t = discLDA(Xtest, means, stds, mu2, covMean, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
        d3t = discLDA(Xtest, means, stds, mu3, covMean, float(nthird)/(nfirst+nsecond+nthird+nfourth))
        d4t = discLDA(Xtest, means, stds, mu4, covMean, float(nfourth)/(nfirst+nsecond+nthird+nfourth))

        predictedTest = np.argmax(np.hstack((d1t, d2t, d3t, d4t)), axis=1)
    else:
        
        
        d1 = discQDA(Xtrain, means, stds, mu1, cov1, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
        d2 = discQDA(Xtrain, means, stds, mu2, cov2, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
        d3 = discQDA(Xtrain, means, stds, mu3, cov3, float(nthird)/(nfirst+nsecond+nthird+nfourth))
        d4 = discQDA(Xtrain, means, stds, mu4, cov4, float(nfourth)/(nfirst+nsecond+nthird+nfourth))

        predictedTrain = np.argmax(np.hstack((d1, d2, d3, d4)), axis=1)

        d1t = discQDA(Xtest, means, stds, mu1, cov1, float(nfirst)/(nfirst+nsecond+nthird+nfourth))
        d2t = discQDA(Xtest, means, stds, mu2, cov2, float(nsecond)/(nfirst+nsecond+nthird+nfourth))
        d3t = discQDA(Xtest, means, stds, mu3, cov3, float(nthird)/(nfirst+nsecond+nthird+nfourth))
        d4t = discQDA(Xtest, means, stds, mu4, cov4, float(nfourth)/(nfirst+nsecond+nthird+nfourth))

        predictedTest = np.argmax(np.hstack((d1t, d2t, d3t, d4t)), axis=1)
        

    for i in range(len(predictedTrain)):
        if predictedTrain[i] == 3:
            predictedTrain[i] = 4
    for i in range(len(predictedTrain)):
        if predictedTrain[i] == 2:
            predictedTrain[i] = 3
    for i in range(len(predictedTrain)):
        if predictedTrain[i] == 1:
            predictedTrain[i] = 2
    for i in range(len(predictedTrain)):
        if predictedTrain[i] == 0:
            predictedTrain[i] = 1
    for i in range(len(predictedTest)):
        if predictedTest[i] == 3:
            predictedTest[i] = 4
    for i in range(len(predictedTest)):
        if predictedTest[i] == 2:
            predictedTest[i] = 3
    for i in range(len(predictedTest)):
        if predictedTest[i] == 1:
            predictedTest[i] = 2
    for i in range(len(predictedTest)):
        if predictedTest[i] == 0:
            predictedTest[i] = 1
    if Type == 1:
        print('LDA Percent correct: Train', percentCorrect(predictedTrain, Ttrain), 'Test', percentCorrect(predictedTest,Ttest))
    else:
        print('QDA Percent correct: Train', percentCorrect(predictedTrain, Ttrain), 'Test', percentCorrect(predictedTest,Ttest))

    plt.plot(Ttrain, 'o-')
    plt.plot(predictedTrain, 'o-')
    plt.title('Ttrain Vs predictedTrain')
    plt.xlabel('samples')
    plt.ylabel('class')
```

```python
def datasplitter(filename,column,Tcolumn):
    f = open(filename,"r")
    data = np.loadtxt(f ,delimiter=',',usecols= column )
    f = open(filename,"r")
    T = np.loadtxt(f ,delimiter=',',usecols= Tcolumn, dtype=np.str).reshape((-1, 1))
    P = np.zeros(len(T)).reshape((-1, 1))
    for i in range(len(T)):  
        T[i] = commandCheck(T[i])
        P[i] = int(np.asscalar(T[i]))
    T = P

    Xnames =  ['sensor1','sensor2','sensor3','sensor4','sensor5','sensor6','sensor7','sensor8','sensor9','sensor10','sensor12','sensor13','sensor14','sensor15','sensor16','sensor17','sensor18','sensor19','sensor20','sensor21','sensor22','sensor23','sensor24']
    X = data
    classes,counts = np.unique(T,return_counts=True)
    Tname = ['Class']
    return X,T,Xnames[column],Tname
```

```python
def classprobplot(X,T,Xnames,Tnames):
    first,_ = np.where(T == 1)
    second,_ = np.where(T == 2)
    third,_ = np.where(T == 3)
    fourth,_ = np.where(T == 4)
    #first = first.reshape((1,-1))
    x1 = X[first]
    x2 = X[second]
    x3 = X[third]
    x4 = X[fourth]
    
    
    means = np.mean(x1, 0)
    stds = np.std(x1, 0)
    # Plot the histogram.
    count, bins, ignored = plt.hist(x1, 30, normed=True)
    plt.plot(bins, 1/(stds * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - means)**2 / (2 * stds**2) ),linewidth=2, color='r')

    means = np.mean(x2, 0)
    stds = np.std(x2, 0)
    # Plot the histogram.
    count, bins, ignored = plt.hist(x2, 30, normed=True)
    plt.plot(bins, 1/(stds * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - means)**2 / (2 * stds**2) ),linewidth=2, color='c')

    means = np.mean(x3, 0)
    stds = np.std(x3, 0)
    # Plot the histogram.
    count, bins, ignored = plt.hist(x3, 30, normed=True)
    plt.plot(bins, 1/(stds * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - means)**2 / (2 * stds**2) ),linewidth=2, color='b')

    means = np.mean(x4, 0)
    stds = np.std(x4, 0)
    # Plot the histogram.
    count, bins, ignored = plt.hist(x4, 30, normed=True)
    plt.plot(bins, 1/(stds * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - means)**2 / (2 * stds**2) ),linewidth=2, color='g')
    
    plt.title('%s Vs %s'%(Xnames,Tnames))
    plt.ylabel("P(Class=k|x)")
    plt.xlabel(Xnames)
    plt.show()
```

```python
def datacomp(filename,Tcolumn,Type):
    f = open(filename,"r")
    data = np.loadtxt(f ,delimiter=',',usecols=0+np.arange(2))
    f = open(filename,"r")
    Tp = np.loadtxt(f ,delimiter=',',usecols= Tcolumn, dtype=np.str).reshape((-1, 1))
    P = np.zeros(len(Tp)).reshape((-1, 1))
    for i in range(len(Tp)):  
        Tp[i] = commandCheck(Tp[i])
        P[i] = int(np.asscalar(Tp[i]))
    Tp = P
    Xp = data

    firstI,_ = np.where(Tp == 1)
    secondI,_ = np.where(Tp == 2)
    thirdI,_ = np.where(Tp == 3)
    fourthI,_ = np.where(Tp == 4)

    firstI = np.random.permutation(firstI)
    secondI = np.random.permutation(secondI)
    thirdI = np.random.permutation(thirdI)
    fourthI = np.random.permutation(fourthI)

    trainFraction = 0.8    
    nfirst = round(trainFraction*len(firstI))
    nsecond = round(trainFraction*len(secondI))
    nthird = round(trainFraction*len(thirdI))
    nfourth = round(trainFraction*len(fourthI))

    rowsTrain = np.hstack((firstI[:nfirst], secondI[:nsecond],thirdI[:nthird], fourthI[:nfourth]))
    Xtrain1 = Xp[rowsTrain]
    Ttrain1 = Tp[rowsTrain, :]
    rowsTest = np.hstack((firstI[nfirst:], secondI[nsecond:], thirdI[nthird:], fourthI[nfourth:]))
    Xtest1 =  Xp[rowsTest]
    Ttest1 =  Tp[rowsTest, :]
    
    if Type == 1:
    
        lda = ql.LDA()
        lda.train(Xtrain1,Ttrain1)
        predictedClasses, probabilities, discriminants = lda.use(Xtest1)
    else:
        qda = ql.QDA()
        qda.train(Xtrain1,Ttrain1)
        predictedClasses, probabilities, discriminants = qda.use(Xtest1)        
        

    plt.figure(figsize=(10, 10))
    plt.subplot(4, 1, 1)
    plt.plot(Xtrain1,Ttrain1,'o', alpha=0.5)
    plt.ylabel("Xtrain vs Train")

    plt.subplot(4, 1, 2)
    plt.plot(Xtest1,probabilities,'o', alpha=0.5)
    if Type == 1:
        plt.ylabel("LDA P(x|Class=k)\n from disc funcs", multialignment="center")
    else:
        plt.ylabel("QDA P(x|Class=k)\n from disc funcs", multialignment="center")
        
    plt.subplot(4, 1, 3)
    plt.plot(Xtest1,discriminants)
    plt.ylabel("discriminants")

    plt.subplot(4, 1, 4)
    plt.plot(Xtest1,predictedClasses,'o', alpha=0.5)
    plt.ylabel("Xtest vs Test")
```

```python
def commandCheck(command):

    if command == "b'Move-Forward'":
        return 1
    if command == "b'Sharp-Right-Turn'":
        return 2
    if command == "b'Slight-Right-Turn'":
        return 3
    if command == "b'Slight-Left-Turn'":
        return 4
```

```python
# code for different folds
def trainQDA(X,T,parameters = 0):
    qda = ql.QDA()
    qda.train(X,T)
    return qda

def evaluateQDA(model,X,T):
    predictedClass,_,_ = model.use(X)   
    return np.sum(predictedClass==T)/X.shape[0] * 100

def trainLDA(X,T,parameters = 0):
    lda = ql.LDA()
    lda.train(X,T)
    return lda

def evaluateLDA(model,X,T):
    predictedClass,_,_ = model.use(X)   
    return np.sum(predictedClass==T)/X.shape[0] * 100

def trainNN(X,T,parameters):
    classes,counts = np.unique(T,return_counts=True)
    nnet = nn.NeuralNetworkClassifier(X.shape[1],parameters[0],len(classes))
    output = nnet.train(X,T,parameters[1], verbose=False)
    return output

def evaluateNN(model,X,T):
    c,Pr,_ = model.use(X,allOutputs=True)    
    return np.sum(c==T)/X.shape[0] * 100

def printResults(label,results):
    print('{:4s} {:>20s}{:>8s}{:>8s}{:>8s}'.format('Algo','Parameters','TrnAcc','ValAcc','TesAcc'))
    print('-------------------------------------------------')
    for row in results:
        # 20 is expected maximum number of characters in printed parameter value list
        print('{:>4s} {:>20s} {:7.2f} {:7.2f} {:7.2f}'.format(label,str(row[0]),*row[1:]))
```

## Conclusion

The soul purpose of this project was to implement the machine learning methods on the simple robot like wall following robot to find the best method to get desired output. In wall following robot, we use ultrasonic sensors which work on the principle of Doppler Effect which is highly nonlinear. So for the analysis purpose, there are 3 datasets with 3 configurations of different no of ultrasonic sensors to better understand the machine learning models. So I took three machine learning methods which include LDA, QDA and neural networks to determine which method works best for the classification of robot output and then implemented these methods on all 3 datasets. By analysing the overall results of my implementation I have concluded following points.
* LDA perform little better when there are 2 sensors but it's performance is worst when we increase the sensors on the robot so this makes it unfit for the wall following robot.
* QDA performed better for the first two datasets but when we increase sensors to 24, QDA performed almost as bad as LDA which makes it unfit for this particular robot problem.
* Neural Networks perform almost perfect on all the datasets so it is the best way to do machine learning when it comes to this particular example of the robot.
* Neural Networks output is little less as compare to first 2 datasets but it can be improved by changing the no of hidden layers,hidden units and iterations.
* Neural Networks takes time in learning when we increase number of hidden layers, hidden units and number of iterations; however, once we get the model, we can just put that model in the machine to take desired decisions which are efficient and also does not take up much memory of the system.<br> 


From this project, I get to know that neural network is the best method for the classification of any nonlinear data although neural network takes too much time to train. The problem I ran into is the time taken by the neural network to train when I increase the no of hidden layers, hidden units, and iterations. Like in the Dataset with 24 sensors, it is giving the accuracy of approximately 92 percentage, so in order to make the accuracy higher, we have to increase the no of hidden layers etc which takes a lot of time.<br>
